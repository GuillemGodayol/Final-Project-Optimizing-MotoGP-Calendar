```python
#Import required libraries
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
```

```python
# Load thd page and get it into so yummy Beautiful Soup
html_doc = 'https://www.motogp.com/en/calendar'
page = requests.get(html_doc)
soup = BeautifulSoup(html_doc, 'html.parser')
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
"""
look for: travelling salesman problem 

phase1: best order for races if they were contiuous
phase2: introduce 'going home' for pilots
phase3: compare distances for pilots to final standings

questions somebody will ask me: where's the analytics part?

library: geopy - per calcular distàncies
""" 
```

```python
print(soup)
```

```python
event_titles = soup.find_all('div', class_ = 'event_title')
```

```python
event_titles
```

```python
titles = []
for event in event_titles:
    temp_list = event.text.split('\n')
    while '' in temp_list:
        temp_list.remove('')
    titles.append(temp_list)
```

```python
titles
```

```python
event_dates = soup.find_all('div', class_ = 'event_date')
```

```python
dates = []
for event in event_dates:
    temp_list = event.text.split('\n\n')
    while '' in temp_list:
        temp_list.remove('')
    
    for i in range(len(temp_list)):
        temp_list[i] = temp_list[i].strip()
    
    dates.append(temp_list)
```

```python
dates
```

```python
races = pd.concat([pd.DataFrame(dates), pd.DataFrame(titles)], axis=1, sort=False)
```

```python
races
```

```python
races.to_csv('../data/races.csv')
```

# Getting Riders info

```python
# Load the page and get the html
riders_url = 'https://www.motogp.com/en/riders/MotoGP'
page = requests.get(riders_url)
riders_html = BeautifulSoup(page.content, 'html.parser')
```

```python
riders_html
```

```python
riders_names = riders_html.find_all('p', class_ = "name")
```

```python
riders_list = []
for rider in riders_names:
    # getting the text of the name:
    name = rider.text
    
    # getting the url of personal page:
    url = riders_html.find_all('a', title = name, limit = 1)[0].get('href')
    
    # surfing into the personal page to get more info:
    page = requests.get(url)
    html = BeautifulSoup(page.content, 'html.parser')
    name = html.find_all('p', class_ = 'name')[0].text
    team = html.find_all('p', class_ = 'team')[0].text
    bike = html.find_all('p', title = re.compile('Bike'))[0].text
    pob = html.find_all('p', title = re.compile('Place'))[0].text
    
    # appending all info to the dict:
    riders_list.append([name, team, bike, pob])
```

```python
riders_list
```

```python
riders = pd.DataFrame(riders_list)
```

```python
riders
```

```python
riders.to_csv('../data/riders.csv')
```
